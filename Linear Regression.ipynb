```python
import numpy as np
def bersihkan_outlier(df, kolom, metode='zscore', threshold=3):
    if metode == 'zscore':
        # Metode Z-score
        z_scores = np.abs((df[kolom] - df[kolom].mean()) / df[kolom].std())
        return df[z_scores < threshold]
    
    elif metode == 'iqr':
        # Metode IQR (Interquartile Range)
        Q1 = df[kolom].quantile(0.25)
        Q3 = df[kolom].quantile(0.75)
        IQR = Q3 - Q1
        
        batas_bawah = Q1 - threshold * IQR
        batas_atas = Q3 + threshold * IQR
        
        return df[(df[kolom] >= batas_bawah) & (df[kolom] <= batas_atas)]

df = bersihkan_outlier(df, kolom_numerik, metode='iqr', threshold=1.5)
df.head()
```

Tujuan:
* Standarisasi nilai fitur numerik dengan menggunakan StandardScaler. Metode ini mengubah setiap nilai numerik dalam dataset menjadi distribusi dengan rata-rata 0 dan standar deviasi 1.
* Kenapa diperlukan?
Banyak algoritma machine learning, seperti regresi logistik, k-NN, dan neural network, lebih sensitif terhadap skala data. Variabel dengan skala besar (misalnya, berat badan dalam kilogram) dapat mendominasi variabel dengan skala kecil (misalnya, tinggi badan dalam meter), sehingga normalisasi diperlukan untuk menyeimbangkan kontribusi masing-masing fitur.

```python
X = df.drop(columns=['BMI'])
y = df['BMI']
```

* X: Berisi semua kolom kecuali kolom BMI (fitur yang digunakan untuk membuat prediksi).

# Linear Regression

Build Model

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

Tujuan:
* Membuat sebuah instance dari model Linear Regression menggunakan class LinearRegression yang disediakan oleh scikit-learn.
* Linear Regression adalah algoritma machine learning yang digunakan untuk menemukan hubungan linear antara variabel input (X) dan output (y). Model ini mencari garis lurus terbaik yang meminimalkan mean squared error (MSE) antara nilai sebenarnya dan nilai prediksi.
* Memasukkan data pelatihan (X_train, y_train) ke dalam model Linear Regression agar model dapat belajar pola antara fitur (X_train) dan target (y_train).

```python
y_pred = linear_model.predict(X_test)
```
